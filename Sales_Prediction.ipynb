```python
X = df[['TV']]
Y = df[['Sales']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

```python
df_1 = LinearRegression()
df_1.fit(X_train, Y_train)
```

```python
X_train_prediction = df_1.predict(X_train)
print(X_train_prediction)
```

```python
print(Y_train)
```

```python
X_test_prediction = df_1.predict(X_test)
print(X_test_prediction)
```

```python
print(Y_test)
```

```python
plt.figure(figsize=(10, 8), facecolor='lightblue', edgecolor='green', alpha=0.5)
plt.scatter(Y_train, X_train_prediction)
plt.xlabel('Original Values')
plt.ylabel('Predicted Values')
plt.title('Original Values vs Predicted Values')
plt.show()
```

```python
plt.figure(figsize=(10, 8), facecolor='lightblue', edgecolor='green', alpha=0.5)
plt.scatter(Y_test, X_test_prediction)
plt.xlabel('Original Values')
plt.ylabel('Predicted Values')
plt.title('Original Values vs Predicted Values')
plt.show()
```
